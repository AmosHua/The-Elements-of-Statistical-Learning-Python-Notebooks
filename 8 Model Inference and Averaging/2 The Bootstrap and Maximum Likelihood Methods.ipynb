```python
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
t = np.array([0., 0., 0., 0., 0.75, 1.5, 2.25, 3., 3., 3., 3.])
```

```python
def B(x, k, i, t):
    if k == 0:
        return 1.0 if t[i] <= x < t[i+1] else 0.0
    if t[i+k] == t[i]:
        c1 = 0.0
    else:
        c1 = (x - t[i])/(t[i+k] - t[i]) * B(x, k-1, i, t)
    if t[i+k+1] == t[i+1]:
        c2 = 0.0
    else:
        c2 = (t[i+k+1] - x)/(t[i+k+1] - t[i+1]) * B(x, k-1, i+1, t)
    return c1 + c2
```

```python
x = np.linspace(0, 3, 400)[:-1]
for n in range(7):
    y = np.zeros(shape=x.shape)
    for i in range(x.shape[0]):
        y[i] = B(x[i], 3, n, t)
    plt.plot(x, y)
```
